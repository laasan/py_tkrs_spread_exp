```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("file20171224.txt",";")
df = df.head(30)
df.head(10)
```

```python
symbol = ((np.array(df['tkr'])).reshape(6,5))
param = ((np.array(df['perfWeek'])).reshape(6,5))
result = df.pivot(index='Yrows',columns='Xcols',values='perfWeek')
labels = (np.asarray(["{0} \n {1:.2f}".format(symb,value)
                     for symb, value in zip(symbol.flatten(),param.flatten())])
         ).reshape(6,5)

fig, ax = plt.subplots(figsize=(12,7))

title = "Somekind heatmap"
plt.title(title,fontsize=18)
ttl = ax.title
ttl.set_position([0.5,1.05])

ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')

sns.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)
plt.show()
```
